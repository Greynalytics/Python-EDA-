# Univariate Analysis

### Aggregate Sales Among Branches

```python
Sales.columns
```

```python
#using Pie plot to visualize the aggregate sales among branches in percentage.
Sales['Branch'].value_counts().plot(kind = 'pie', autopct = '%1.2f%%', shadow = True)
plt.show()
```

```python
sns.countplot(Sales['Branch'])
plt.title('Aggregate Sales Among Three branches')
plt.show()
#used countplot in seaborn library to visualize the plot.
#from both plot we can infer that branch A has the highest
```

### Most popular payment method used by customer

```python
Sales['Payment'].value_counts()
```

```python
Sales['Payment'].value_counts().plot(kind ='bar')
plt.title('Most Popular Payment Method Used by Customer')
plt.xlabel('Payment Methods')
plt.show()
```

 ### Distribution of various Customer Rating 

```python
Sales['Rating'].value_counts()
```

```python
sns.distplot(Sales['Rating'])
plt.title('Distribution of Customer Rating')
plt.show()
#Used displot in python library to show the distribution of customer rating.
#from the plot below we can suggest that the rating data is normally distributed
#no skewness in the distribution
```

```python
#we can further check the skewness
#distribution of rating is zero which indicated no skewness in the dataset

Sales['Rating'].skew
```

### Distribution of COGS - Cost of Goods Sold

```python
sns.distplot(Sales['cogs'])
plt.show()
#the distribution curve for cost of goods sold was skewed to 
#the right which may indicate a positive skewness 
```

```python
Sales['cogs'].skew() #this further affirm the above curve 
#which show right skewness giving a positive value of 0.892
```

# Bivariate and Multivariate Analysis - comparing two or more variables

### Is cost of goods sold affected by the Rating that customers provides

```python
sns.scatterplot(x = 'cogs', y = 'Rating',data = Sales)
plt.title('Relationship between COGS and Customer Rating')
plt.xlabel('Cost of Goods Sold')
plt.ylabel('Customer Rating')
plt.show()
#below scatterplot fig infer that, we have no relationship between 
#cost of goods sold and ratings customers provide
```

### Does Gross Income affects the Rating that customers provides

```python
sns.scatterplot(x = 'gross income', y = 'Rating', data = Sales)
plt.title('Relationship between Gross Income and Customer Rating')
plt.xlabel('Gross Income')
plt.ylabel('Customer Rating')
plt.show()
```

### Most profitable branch as per Gross Income by city 

```python
sns.barplot(x = 'Branch', y = 'gross income', data = Sales, hue = 'City')
plt.title('Most Profitable Branch per Gross Profit')
plt.xlabel('Branch')
plt.ylabel('gross income', fontsize = 12)
plt.show()
```

### productline generating the most Income

```python
sns.barplot(x = 'Product line', y = 'gross income', data = Sales)
plt.xticks(rotation = 80)
plt.title('Product line with the most Gross Income')
plt.xlabel('Product Line')
plt.ylabel('Gross Income')
plt.show()
```

### Highest Unit price in the product Line

```python
sns.barplot(x = 'Product line', y = 'Unit price', data = Sales)
plt.xticks(rotation = 80)
plt.title('Product line with the Highest unit price')
plt.xlabel('Product Line')
plt.ylabel('Unit Price')
plt.show()
```

### Different payment methods used by customer citywise

```python
pd.crosstab(Sales['City'], Sales['Payment'])
```

```python
sns.heatmap(pd.crosstab(Sales['City'], Sales['Payment']))
```

### Product Line purchased in the highest quantity

```python
Sales.groupby('Product line').sum()['Quantity'].plot(kind = 'bar')
plt.title('Product line with the Highest Quantity')
plt.xlabel('Product Line')
plt.ylabel('Quantity')
plt.show()
```
